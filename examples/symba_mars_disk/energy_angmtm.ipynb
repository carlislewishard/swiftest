```python
import numpy as np
import pandas as pd
```

```python
egy = pd.read_csv("energy.out")
```

```python
egy['(Eorbit - E0)/E0'] = (egy['Eorbit'] - egy['Eorbit'][0]) / np.abs(egy['Eorbit'][0])
ax = egy['(Eorbit - E0)/E0'].plot(x='t')
ax.set_xlabel("t (s)")
ax.set_ylabel("$(E_{orbit} - E_0)/E_0$")
ax.set_xscale('log')
```

```python
egy['(Eorbit + Ecollisions - E0)/E0'] = ( egy['Eorbit'] + egy['Ecollisions'] - egy['Eorbit'][0]) / egy['Eorbit'][0]
ax = egy['(Eorbit + Ecollisions - E0)/E0'].plot(x='t')
ax.set_xlabel("t (s)")
ax.set_ylabel("$(E_{orbit} + E_{collisions} - E_0)/E_0$")
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
egy['Ltot'] = np.sqrt(egy['Lx']**2 + egy['Ly']**2 + egy['Lz']**2)
egy['|(L - L0)/L0|'] = np.abs((egy['Ltot'] - egy['Ltot'][0]) / egy['Ltot'])
ax = egy['|(L - L0)/L0|'].plot(x='t')
ax.set_xlabel("t (s)")
ax.set_ylabel('$|(L - L_0)/L_0|$')
ax.set_xscale('log')
ax.set_yscale('log')
```

```python

egy['|(M - M0)/M0|'] = np.abs((egy['Mtot'] - egy['Mtot'][0]) / egy['Mtot'])
ax = egy['|(M - M0)/M0|'].plot(x='t')
ax.set_xlabel("t (s)")
ax.set_ylabel('$|(M - M_0)/M_0|$')
ax.set_xscale('log')
ax.set_yscale('log')
```
