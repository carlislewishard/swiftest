```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
egy = pd.read_csv("energy.out")
```

```python
egy['Ltot'] = np.sqrt(egy['Lx']**2 + egy['Ly']**2 + egy['Lz']**2)
egy['Etot'] = egy['Eorbit'] + egy['Ecollisions']
egy['(Eorbit - E0)/|E0|'] = (egy['Eorbit'] - egy['Eorbit'][0]) / np.abs(egy['Eorbit'][0])
egy['Ecollisions/|E0|'] = egy['Ecollisions'] / np.abs(egy['Eorbit'][0])
egy['(Etot - E0)/|E0|'] = (egy['Etot'] - egy['Eorbit'][0]) / np.abs(egy['Eorbit'][0])
egy['(L - L0)/L0'] = (egy['Ltot'] - egy['Ltot'][0]) / egy['Ltot']
egy['|(M - M0)/M0|'] = np.abs((egy['Mtot'] - egy['Mtot'][0]) / egy['Mtot'])
```

```python
ax = egy['(Etot - E0)/|E0|'].plot(x='t')
ax.set_xlabel("t (s)")
ax.set_ylabel("$(E_{tot} - E_0)/|E_0|$")
```

```python
ax = egy.plot(x='t', y='Ecollisions/|E0|')
ax.set_xlabel("t (s)")
ax.set_ylabel("$'Ecollisions/|E0|'$")
ax.set_yscale('log')
#ax.set_ylim([-1e-9, 1e-3])
plt.savefig("Ecollision.png")
```

```python
egy['|(L - L0)/L0|'] = np.abs((egy['Ltot'] - egy['Ltot'][0]) / egy['Ltot'])
ax = egy.plot(x='t', y='|(L - L0)/L0|')
ax.set_xlabel("t (s)")
ax.set_ylabel('$|(L - L_0)/L_0|$')
#ax.set_yscale('symlog')
ax.set_yscale('log')
ax.set_ylim([1e-16, 1e-12])
plt.savefig("AngularMomentum.png")
```
